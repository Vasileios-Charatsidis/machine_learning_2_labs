### Notation

$\mathbf{X}$ is the $M \times T$ data matrix, containing $M$ measurements at $T$ time steps.

$\mathbf{S}$ is the $S \times T$ source matrix, containing $S$ source signal values at $T$ time steps. We will assume $S = M$.

$\mathbf{A}$ is the mixing matrix. We have $\mathbf{X} = \mathbf{A S}$.

$\mathbf{W}$ is the matrix we aim to learn. It is the inverse of $\mathbf{A}$, up to indeterminacies (scaling and permutation of sources).

$\phi$ is an elementwise non-linearity or activation function, typically applied to elements of $\mathbf{W X}$.

````python
%pylab inline
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'Arial'})
from scipy.special import *
from scipy import integrate
import sys
# reload(sys)
# sys.setdefaultencoding("utf-8") # because notebooks are retarded

# Signal generators
def sawtooth(x, period=0.2, amp=1.0, phase=0.):
    return (((x / period - phase - 0.5) % 1) - 0.5) * 2 * amp

def sine_wave(x, period=0.2, amp=1.0, phase=0.):
    return np.sin((x / period - phase) * 2 * np.pi) * amp

def square_wave(x, period=0.2, amp=1.0, phase=0.):
    return ((np.floor(2 * x / period - 2 * phase - 1) % 2 == 0).astype(float) - 0.5) * 2 * amp

def triangle_wave(x, period=0.2, amp=1.0, phase=0.):
    return (sawtooth(x, period, 1., phase) * square_wave(x, period, 1., phase) + 0.5) * 2 * amp

def random_nonsingular_matrix(d=2):
    """
    Generates a random nonsingular (invertible) matrix of shape d*d
    """
    epsilon = 0.1
    A = np.random.rand(d, d)
    while abs(np.linalg.det(A)) < epsilon:
        A = np.random.rand(d, d)
    return A

def plot_signals(X, title=""):
    """
    Plot the signals contained in the rows of X.
    """
    figure(figsize=(8,8))
    for i in range(X.shape[0]):
        ax = plt.subplot(X.shape[0], 1, i + 1)
        plot(X[i, :])
        ax.set_xticks([])
        ax.set_yticks([])
    plt.suptitle(title, fontsize=20)
    plt.show()
````

### 1.3 Implicit priors (20 points)
As explained in MacKay's book, an activation function $\phi$ used in the ICA learning algorithm corresponds to a prior distribution over sources. Specifically, $\phi(a) = \frac{d}{da} \ln p(a)$. For each of the following activation functions, derive the source distribution they correspond to.
$$\phi_0(a) = -\tanh(a)$$
$$\phi_1(a) = -a + \tanh(a)$$
$$\phi_2(a) = -a^3$$
$$\phi_3(a) = -\frac{6a}{a^2 + 5}$$

The normalizing constant is not required, so an answer of the form $p(a) \propto \verb+[answer]+$ is ok.

Plot the activation functions and the corresponding prior distributions. Compare the shape of the priors to the histogram you plotted in the last question.

### Solution

Given the differential equation $\frac{d}{da} \ln p(a) = \phi(a)$, the general solution is given by $p(a) = k \exp \left\{ \int \phi(a) da\right\}$, where $k$ is a constant such that $\int_{\Omega} p(a) da = 1$ with $\Omega$ being the domain of $a$. We computed the integration constants by hand, however we omit this calculations for brevity. If required, please let us know. Below, $c$ and $c_i$ represent integration constants.

#### Case $\phi_0(a) = -\tanh(a)$
$$ \phi_0(a) = - \frac{\sinh(a)}{\cosh(a)} = - \frac{e^a - e^{-a}}{e^a + e^{-a}} $$

Let $u = e^a + e^{-a}$ such that $du = (e^a - e^{-a}) da$

$$ \int - \frac{e^a - e^{-a}}{e^a + e^{-a}} da = - \int \frac{1}{u} du = - \log \lvert u \lvert + c = - \log \lvert 2 \cosh(a) \lvert + c = \log \left( \frac{c}{\cosh(a)} \right) \hspace{5mm} \rightarrow \hspace{5mm} p_0(a) = \frac{1}{\pi \cosh(a)}$$

#### Case $\phi_1(a) = -a + \tanh(a)$

We use the result from above. 

$$\int -a + \tanh(a) da = \frac{-a^2}{2} - \log \left( \frac{c_1}{\cosh(a)} \right) + c_2 \hspace{5mm} \rightarrow \hspace{5mm}  p_1(a) = \frac{1}{\sqrt{ 2\pi e}} \exp \left( \frac{-a^2}{2} \right) \cosh(a) $$

#### Case $\phi_2(a) = -a^3$

$$ \int -a^3 da = \frac{-a^4}{4} \hspace{5mm} \rightarrow \hspace{5mm} p_2(a) = \frac{\sqrt{2}}{\Gamma(\frac{1}{4})} \exp \left( \frac{-a^4}{4} \right) $$

#### Case $\phi_3(a) = -\frac{6a}{a^2 + 5}$

Let $u = u^2 + 5$ such that $du = 2a \, da$.

$$ \int-\frac{6a}{a^2 + 5} da = \int \frac{-3}{u} du = - 3 \log \lvert u \lvert +c = - 3 \log \lvert a^2 + 5 \lvert +c \hspace{5mm} \rightarrow \hspace{5mm} p_3(a) = \frac{200 \sqrt{5}}{3 \pi} (a^2 + 5) ^{-3}$$

#### Comparison

For the random signal, none of the priors are appropriate because the source is originally a Gaussian distribution, whereas all of the priors are either sub- or super-Gaussian (see section on Kurtosis at the bottom of this report). For the sawtooth and triangle signals, due to their uniform-like distributions, the $p_2$ prior with an appropriate gain could be suitable. For the cosine and square signals, none of the priors are appropriate due to the bimodality of the source distributions. In the case of the square signal, an activation function specified as $\phi(x) = \delta_{-1}(x) + \delta_1(x)$ would be appropriate.

````python
### 1.3 Implicit priors (continued)
activation_fx_names = ['Phi 0', 'Phi 1', 'Phi 2', 'Phi 3']

def activation(x, case):
    return {
        0: - np.tanh(x),
        1: -x + np.tanh(x),
        2: -x**3,
        3: -6*x/(x**2+5),
    }[case]

def prior(x, case):
    return {
        0: (1/np.pi) * 1/np.cosh(x),
        1: (1/np.sqrt(2*np.pi*np.exp(1))) * np.exp(-(x**2)/2) * np.cosh(x),
        2: (np.sqrt(2)/gamma(1./4)) * np.exp(-(x**4)/4),
        3: (200*np.sqrt(5))/(3 * np.pi) * (x**2 + 5)**(-3),
    }[case]

# Sample activation functions
x = np.linspace(-1.5,1.5,100)
activ_data = np.array([activation(x,0), activation(x,1), activation(x,2), activation(x,3)])

# Plot activation data
plt.plot(x, activ_data.T)
plt.legend(['$\phi_0$', '$\phi_1$', '$\phi_2$', '$\phi_3$'])
plt.show()

# Sample prior functions
x = np.linspace(-5,5,100)
prior_data = np.array([prior(x,0), prior(x,1), prior(x,2), prior(x,3)])

# Plot prior data
plt.plot(x, prior_data.T)
plt.legend(['$p_0$', '$p_1$', '$p_2$', '$p_3$'])
plt.show()

# Verify that normalization constants are correct
print('Normalization constant verification:')
for i in range(4):
    prior_fn = lambda x: prior(x,i)
    res = integrate.quad(prior_fn, -100, 100)
    print('   Prior %d: Integral: %f - Error: %s' % (i , res[0], res[1]))
````

### 1.4 Whitening (15 points)
Some ICA algorithms can only learn from whitened data. Write a method `whiten(X)` that takes a $M \times T$ data matrix $\mathbf{X}$ (where $M$ is the dimensionality and $T$ the number of examples) and returns a whitened matrix. If you forgot what whitening is or how to compute it, various good sources are available online, such as http://courses.media.mit.edu/2010fall/mas622j/whiten.pdf

````python
### 1.4 Whitening

def whiten(X):
    cov = np.cov(X)
    Lambda, Phi = np.linalg.eig(cov)

    Y = Phi.T.dot(X)
    L_sqrt = np.sqrt(np.diag(1/Lambda))

    W = L_sqrt.dot(Y)
    return W
    
Xw = whiten(X)
````

### 1.7 Independent Component Analysis (25 points)
Implement the covariant ICA algorithm as described in MacKay. Write a function `ICA(X, activation_function, learning_rate)`, that returns the demixing matrix $\mathbf{W}$. The input `activation_function` should accept a function such as `lambda a: -tanh(a)`. Update the gradient in batch mode, averaging the gradients over the whole dataset for each update. Try to make it efficient, i.e. use matrix operations instead of loops where possible (loops are slow in interpreted languages such as python and matlab, whereas matrix operations are internally computed using fast C code).

````python
### 1.7 Independent Component Analysis

def ICA(X, activation_function, learning_rate, S=None, epochs=1000, verbose = False, batch_size=0, num_evaluations=20):

    D,N = X.shape
    W = random_nonsingular_matrix(d=X.shape[0]) # initialize W
    I = np.eye(D)
    
    # Batch size
    if batch_size == 0:
        batch_size = N
    
    for epoch in range(epochs):
        rand_samples = numpy.random.choice(N, batch_size) # randomly sample with replacement
        X_batch = X[:,rand_samples]
        
        a = W.dot(X_batch)
        W += (learning_rate / batch_size) * (I+ activation_function(a).dot(a.T)).dot(W) # update rule
        
    return W
````

### 1.10 Excess Kurtosis (15 points)
The (excess) kurtosis is a measure of 'peakedness' of a distribution. It is defined as
$$
\verb+Kurt+[X] = \frac{\mu_4}{\sigma^4} - 3 = \frac{\operatorname{E}[(X-{\mu})^4]}{(\operatorname{E}[(X-{\mu})^2])^2} - 3
$$
Here, $\mu_4$ is known as the fourth moment about the mean, and $\sigma$ is the standard deviation.
The '-3' term is introduced so that a Gaussian random variable has 0 excess kurtosis.
We will now try to understand the performance of the various activation functions by considering the kurtosis of the corresponding priors, and comparing those to the empirical kurtosis of our data.

#### 1.10.1 (10 points)
First, compute analytically the kurtosis of the four priors that you derived from the activation functions before. To do this, you will need the normalizing constant of the distribution, which you can either obtain analytically (good practice!), using computer algebra software (e.g. Sage) or by numerical integration (see scipy.integrate).

### Solution

Note that all the priors are symmetric with respect to 0, so $\mu = 0$.
$$\verb+Kurt+[X] = \frac{\mu_4}{\sigma^4} - 3 = \frac{\operatorname{E}[X^4]}{\operatorname{E}[X^2]^2} - 3$$

#### Case $p_0(a) = \frac{1}{\pi \cosh(a)}$

$$ \verb+Kurt+[X] = \frac{\int_{R} \frac{1}{\pi \cosh(x)} x^4 dx }{\left( \int_{R} \frac{1}{\pi \cosh(x)} x^2 dx \right)^2} - 3 = \frac{\frac{5 \pi^4}{16}} {\frac{\pi^4}{16}} - 3 = 2$$

#### Case $p_1(a) = \frac{1}{\sqrt{ 2\pi e}} \exp \left( \frac{-a^2}{2} \right) \cosh(a)$

$$ \verb+Kurt+[X] = \frac{\int_{R} \frac{1}{\sqrt{ 2\pi e}} \exp \left( \frac{-x^2}{2} \right) \cosh(x) x^4 dx }{\left( \int_{R} \frac{1}{\sqrt{ 2\pi e}} \exp \left( \frac{-x^2}{2} \right) \cosh(x) x^2 dx \right)^2} - 3 = \frac{10}{4} - 3 = -0.5$$

#### Case $p_2(a) = \frac{\sqrt{2}}{\Gamma(\frac{1}{4})} \exp \left( \frac{-a^4}{4} \right)$

$$ \verb+Kurt+[X] = \frac{\int_{R} \frac{\sqrt{2}}{\Gamma(\frac{1}{4})} \exp \left( \frac{-x^4}{4} \right) x^4 dx }{\left( \int_{R} \frac{\sqrt{2}}{\Gamma(\frac{1}{4})} \exp \left( \frac{-x^4}{4} \right) x^2 dx \right)^2} - 3 = \frac{1} {0.45694} - 3 = -0.8115$$

#### Case $p_3(a) = \frac{200 \sqrt{5}}{3 \pi} (a^2 + 5) ^{-3}$

$$ \verb+Kurt+[X] = \frac{\int_{R} \frac{200 \sqrt{5}}{3 \pi} (x^2 + 5) ^{-3} x^4 dx }{\left( \int_{R} \frac{200 \sqrt{5}}{3 \pi} (x^2 + 5) ^{-3} x^2 dx \right)^2} - 3 = \frac{225} {25} - 3 = 6$$


Below, you can see the numerical integrations that support this values. Note that these are numerical approximations to improper integrals.

````python
def mu_k(x, case, k):
    return {
        0: (1/np.pi) * 1/np.cosh(x) * x**k,
        1: (1/np.sqrt(2*np.pi*np.exp(1))) * np.exp(-(x**2)/2) * np.cosh(x)  * x**k,
        2: (np.sqrt(2)/gamma(1.0/4)) * np.exp(-(x**4)/4.0)  * x**k,
        3: (200*np.sqrt(5))/(3 * np.pi) * (x**2 + 5)**(-3) * x**k,
    }[case]

lim = 700

# Verify that normalization constants are correct
print('Theoretical Kurtosis Calculation:')
for i in range(4):
    mu4_fn = lambda x: mu_k(x,i,4)
    mu2_fn = lambda x: mu_k(x,i,2)
    mu4 = integrate.quad(mu4_fn, 0, lim, limit=800)
    mu2 = integrate.quad(mu2_fn, 0, lim, limit=800)
    print('   Prior %d: Kurt: %f' % (i , mu4[0]/(2*mu2[0]**2) - 3))
````
